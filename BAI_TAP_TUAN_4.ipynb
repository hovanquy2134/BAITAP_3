```python
import numpy as np
import pandas as pd
import joblib
import ipywidgets as widgets
from IPython.display import display, clear_output

from sklearn import datasets
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

data = pd.read_csv('/content/diabetes.csv', sep=';')
x = data.loc[:,['Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
y = data.loc[:,'Outcome']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)


sc = StandardScaler()
sc.fit(x_train)
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)

model = Perceptron(max_iter=1000,eta0=0.05,random_state = 42)
model.fit(x_train_std,y_train)
joblib.dump(model,'model.pkl')
joblib.dump(sc,'scaler.pkl')

Glu_widget = widgets.IntSlider(value=30,min =0,max=199,step=1,description='Glucose')
BP_widget = widgets.IntSlider(value=30,min =0,max=122,step=1,description='BloodPressure')
ST_widget = widgets.IntSlider(value=30,min =0,max=99,step=1,description='SkinThickness')
In_widget = widgets.IntSlider(value=30,min =0,max=846,step=1,description='Insulin')
BMI_widget = widgets.FloatSlider(value=30,min =0,max=67.1,step=0.1,description='BMI')
DPF_widget = widgets.FloatSlider(value=1,min =0.078,max=2.42,step=0.01,description='DiabetesPedigreeFunction')
Age_widget = widgets.IntSlider(value=30,min =21,max=81,step=1,description='Age')

button = widgets.Button(description="Predict", button_style="success")
output = widgets.Output()

def predict(b):
  with output:
    clear_output()

    x_new = np.array([[Glu_widget.value,BP_widget.value,ST_widget.value,In_widget.value,BMI_widget.value,DPF_widget.value,Age_widget.value]])
    x_new_std = sc.transform(x_new)
    prediction = model.predict(x_new_std)

    if prediction[0] == 1:
      print('bị tiểu đường rồi bạn ơi!!!')
    else:
      print('không bị tiểu đường rồi bạn ơi!!!')

button.on_click(predict)

display(Glu_widget,BP_widget,ST_widget,In_widget,BMI_widget,DPF_widget,Age_widget,button,output)

```
